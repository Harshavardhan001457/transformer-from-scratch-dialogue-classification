```python
import pandas as pd
import re
import json
```

```python
datapath1 = './DailyDialog/train.csv'
datapath2 = './DailyDialog/validation.csv'
datapath3 = './DailyDialog/test.csv'
```

```python
df1 = pd.read_csv(datapath1)
df2 = pd.read_csv(datapath2)
df3 = pd.read_csv(datapath3)
```

```python
def parse_dialog(text):
    try:
        result = re.findall(r"""(['"])(.*?)(?<!\\)\1""", text, re.DOTALL)
        result = [s.strip() for _, s in result]
        return result
    except Exception:
        pass
    return None
```

```python
def preprocess(df):
  LABEL_TYPE = "emotion"
  dialogues = []
  labels = []

  for _, row in df.iterrows():
      dialog = parse_dialog(row["dialog"])

      text = row[LABEL_TYPE][1:-1]
      label_seq = text.split(' ')
      label_seq = [int(i) for i in label_seq]
      if dialog is None:
          continue

      if len(dialog) != len(label_seq):
          continue

      dialogues.append(dialog)
      labels.append(label_seq)
  data = {"dialogues": dialogues, "labels": labels}
  return data
```

```python
train_data = preprocess(df1)
val_data = preprocess(df2)
test_data = preprocess(df3)
```

```python
with open('./train_data.json', "w", encoding="utf-8") as f:
    json.dump(train_data,f,indent=2)
with open('./val_data.json', "w", encoding="utf-8") as f:
    json.dump(val_data,f,indent=2)
with open('./test_data.json', "w", encoding="utf-8") as f:
    json.dump(test_data,f,indent=2)
```
